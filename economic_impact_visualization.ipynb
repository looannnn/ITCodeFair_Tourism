# Economic Impact Score Weighting Visualization

This notebook creates a pie chart to illustrate the weighting components of the economic impact composite score used in the tourism analysis model.

## Import Required Libraries

Import matplotlib.pyplot for plotting and any other necessary libraries for data visualization.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Set style for better looking plots
plt.style.use('default')
sns.set_palette("husl")
```

## Define Composite Score Components

Define the components and their respective weights: rating (40%), category weight (25%), review length (15%), emotion (15%), and specific aspects (5%).

```python
# Define the economic impact score components and their weights
components = {
    'Rating Score': 40,
    'Category Weight': 25, 
    'Text Engagement (Length)': 15,
    'Emotional Intensity': 15,
    'Aspect Specificity': 5
}

# Display the components
print("Economic Impact Score Components:")
print("=" * 40)
for component, weight in components.items():
    print(f"{component}: {weight}%")
    
print(f"\nTotal: {sum(components.values())}%")
```

## Create Pie Chart Data

Organize the component names and weights into lists or arrays suitable for pie chart plotting.

```python
# Extract labels and sizes for pie chart
labels = list(components.keys())
sizes = list(components.values())

# Define colors for each component (professional color scheme)
colors = [
    '#FF6B6B',  # Rating Score - Red
    '#4ECDC4',  # Category Weight - Teal  
    '#45B7D1',  # Text Engagement - Blue
    '#FFA07A',  # Emotional Intensity - Orange
    '#98D8C8'   # Aspect Specificity - Green
]

print("Pie Chart Data:")
print(f"Labels: {labels}")
print(f"Sizes: {sizes}")
print(f"Colors: {colors}")
```

## Generate Basic Pie Chart

Create a basic pie chart using matplotlib's pie() function with the defined data.

```python
# Create basic pie chart
plt.figure(figsize=(10, 8))

# Basic pie chart
plt.pie(sizes, labels=labels, colors=colors, startangle=90)
plt.title('Economic Impact Score Components - Basic Version', fontsize=16, fontweight='bold', pad=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

## Customize Pie Chart Appearance

Enhance the pie chart with colors, explode effects for emphasis, and styling options to improve visual appeal.

```python
# Create enhanced pie chart with explode effect and better styling
plt.figure(figsize=(12, 10))

# Explode the largest slice (Rating Score) to emphasize it
explode = (0.1, 0.05, 0, 0, 0)  # explode 1st slice (Rating Score) and slightly the 2nd

# Create pie chart with enhanced styling
wedges, texts, autotexts = plt.pie(
    sizes, 
    labels=labels, 
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    textprops={'fontsize': 11, 'fontweight': 'bold'}
)

# Enhance the appearance
plt.title('Economic Impact Score Component Weighting\n(Tourism Experience Analysis)', 
          fontsize=18, fontweight='bold', pad=30)

# Style the percentage text
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(12)

plt.axis('equal')
plt.show()
```

## Add Percentage Labels and Legend

Add percentage labels to each slice and include a legend for better readability and interpretation.

```python
# Create pie chart with legend and detailed labels
fig, ax = plt.subplots(figsize=(14, 10))

# Create pie chart
wedges, texts, autotexts = ax.pie(
    sizes, 
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    textprops={'fontsize': 12, 'fontweight': 'bold'},
    pctdistance=0.85
)

# Add title
ax.set_title('Economic Impact Score Component Weighting\n(Tourism Experience Analysis)', 
             fontsize=20, fontweight='bold', pad=30)

# Style the percentage text
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold')
    autotext.set_fontsize(13)

# Create legend with component descriptions
legend_labels = [
    f'Rating Score ({sizes[0]}%)\nUser rating impact',
    f'Category Weight ({sizes[1]}%)\nBusiness type value',
    f'Text Engagement ({sizes[2]}%)\nReview detail level', 
    f'Emotional Intensity ({sizes[3]}%)\nEmotional response',
    f'Aspect Specificity ({sizes[4]}%)\nSpecific mentions'
]

ax.legend(wedges, legend_labels,
          title="Component Details",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1),
          fontsize=11)

plt.axis('equal')
plt.tight_layout()
plt.show()
```

## Display Final Visualization

Display the completed pie chart with title and proper formatting to illustrate the economic impact score weighting.

```python
# Create final professional pie chart for presentation
plt.figure(figsize=(15, 12))

# Create the pie chart with all enhancements
wedges, texts, autotexts = plt.pie(
    sizes, 
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    textprops={'fontsize': 14, 'fontweight': 'bold'},
    pctdistance=0.8,
    labeldistance=1.1
)

# Main title and subtitle
plt.suptitle('Economic Impact Score Composition', fontsize=24, fontweight='bold', y=0.95)
plt.title('Weighting of Components in Tourism Experience Valuation Model', 
          fontsize=16, style='italic', pad=20)

# Style the percentage text inside slices
for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontweight('bold') 
    autotext.set_fontsize(14)

# Add explanation text
explanation_text = """
Key Insights:
• Rating Score (40%): Primary indicator of user satisfaction
• Category Weight (25%): Business type economic multiplier  
• Text Engagement (15%): Review depth and engagement level
• Emotional Intensity (15%): Emotional response strength
• Aspect Specificity (5%): Detailed aspect mentions
"""

plt.figtext(0.02, 0.15, explanation_text, fontsize=12, 
           bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgray", alpha=0.8))

# Equal aspect ratio and layout
plt.axis('equal')
plt.tight_layout()

# Save the chart
plt.savefig('economic_impact_score_weighting.png', dpi=300, bbox_inches='tight')
plt.show()

print("✅ Final visualization completed and saved as 'economic_impact_score_weighting.png'")
```
